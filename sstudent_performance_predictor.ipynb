```python
df['alc'] = (df['Dalc'] + df['Walc']) / 2
```

```python
df.drop(columns=['Dalc','Walc','famrel','famsize','address','school',
    'sex','Pstatus','Mjob','Fjob','reason','guardian',
    'schoolsup','famsup','paid','activities','nursery','romantic'],axis=1,inplace=True)
```

```python
df.info()
```

```python
df['internet'].unique()
```

```python
df['health'].unique()
```

```python
df['internet'] = df['internet'].map({'yes':1,'no':0})
df['higher'] = df['higher'].map({'yes':1,'no':0})

df = pd.get_dummies(df,drop_first=True)
```

```python
X = df.drop('G3',axis=1)
y = df['G3']
```

```python
X
```

```python
df.info()
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test ,y_train , y_test = train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
model = RandomForestRegressor(n_estimators=100,max_depth=5,random_state=42)
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
mse = mean_squared_error(y_pred,y_test)
np.sqrt(mse)
```

```python
joblib.dump(model,"model.pkl")
```
